```python
import glob

import pandas as pd

# read dataset splits from ../MissingDataOpenData/data_splits
train_df = pd.read_csv("../MissingDataOpenData/data_splits/training.txt", header=None)
val_df = pd.read_csv("../MissingDataOpenData/data_splits/validation_200.txt", header=None)
test200_df = pd.read_csv("../MissingDataOpenData/data_splits/final_test_500.txt", header=None)

# get all the original images from ../MissingDataOpenData/originals
originals = glob.glob("../MissingDataOpenData/originals/*.jpg")
masked = glob.glob("../MissingDataOpenData/masked/*.png")
masks = glob.glob("../MissingDataOpenData/masks/*.png")
target_folder = "final_test_500"

!mkdir -p cat_dataset/{target_folder}
n=0

for _,img_path in enumerate(masked):
    # extract image id from path
    img_id = '_'.join(img_path.split("/")[-1].split(".")[0].split("_")[:2])
    #print(img_id)
    # create folder for image in cat_dataset
    
    if img_id in test200_df.values:
        # copy image to ../MissingDataOpenData/data_splits/train
        n += 1

        !cp {img_path} cat_dataset/{target_folder}/{img_id}.png
        print(f"copied {img_path} to cat_dataset/{target_folder}")

for _,img_path in enumerate(masks):
    # extract image id from path
    img_id = '_'.join(img_path.split("/")[-1].split(".")[0].split("_")[:2])
    #print(img_id)
    # create folder for image in cat_dataset
    
    if img_id in test200_df.values:
        # copy image to ../MissingDataOpenData/data_splits/train
        n += 1

        !cp {img_path} cat_dataset/{target_folder}/{img_id}_mask.png
        print(f"copied {img_path} to cat_dataset/{target_folder}")
```
